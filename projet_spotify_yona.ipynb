```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/natihuri1/spotifyProject/main/spotifyProject.csv"
spotify_df = pd.read_csv(url)
```

```python
spotify_df
```

```python
spotify_df.isnull().sum()
```

```python
spotify_df.hist(bins=100)
```

```python
popular_above_80 = spotify_df[spotify_df.popularity > 80]
not_popular = spotify_df[(spotify_df.popularity < 80) & (spotify_df.popularity > 20)]
popular_above_80.hist(bins=10)
```

```python
not_popular.hist(bins=10)
```

```python
features = ['acousticness', 'danceability','energy','instrumentalness','speechiness', 'tempo', 'valence']
fig, axes = plt.subplots(3,3,figsize=(15,15))
plt.subplots_adjust(wspace=0.3, hspace = 0.3)
axes = axes.flatten()
for i,att in enumerate(features):
    sns.histplot( x=att , y = 'popularity', data=spotify_df ,  ax=axes[i] )
```

```python
df_to_plot = spotify_df.drop(columns=['popularity','duration_ms','liveness','loudness'])
df_to_plot.plot(subplots=True, layout=(5,4), kind='box', figsize=(12,14), patch_artist=True)
plt.subplots_adjust(wspace=0.5)
```

```python
features = ['acousticness', 'danceability','energy','instrumentalness','speechiness', 'tempo', 'valence','popularity']

spotify_df[features].plot.density(subplots=True, sharex=False, sharey=False, layout=(3, 3), figsize=(12,14))
```
